### Challenge 2: Count Commits

Count how many commits were made in the past week.

1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

1. Count how many commit objects are contained in the array

```python
api_key="token"
print(os.getenv("token"))
```

```python
def requestGithub(url):
    token = os.getenv("token")
    if not token:
        raise ValueError ("Necesitas un token")
    print(f"Haciendo request a {url}")
    
    params = {"Auhorization" : f"token {api_key}"}
    response = requests.get( url = url , params = params).json()
    
    return response
url="https://api.github.com/repos/ironhack-datalabs/datamad1020-rev/commits"
```

```python
respuesta = requestGithub(url)
print(respuesta)
```

```python
commits=[elemento['commit'] for elemento in respuesta]
len(commits)
```
